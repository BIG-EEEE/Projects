```python
# importing required libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

## DATA ASSIMILATION WITH OPTIMAL INTERPOLATION

```python
#loading observation data( in degree celcius)
obs= np.array([15.2, 16.1, 14.5, 15.8, 25.0 ])

#loading model_output data( in degree celcius)
model_output= np.array([14.8, 15.5, 14.0, 16.0, 25.3 ])

    #-------PERFORMING OPTIMAL INTERPOLATION------#

# 1. ---------Calculating weights of both dataset-------
# calculating variances
var_obs = np.var(obs)
var_mod = np.var(model_output)
# Calculating for required weights
W_obs = 1 / var_obs / (1 / var_obs + 1 / var_mod)
W_mod = 1/ var_mod / (1 / var_obs + 1 / var_mod)

# 2. Calculating Optimal Interpolation
analyzed_data = W_obs*obs + W_mod*model_output

#3. Calculating root mean squared error(rsme)
rmse = np.sqrt(np.mean((analyzed_data-obs)**2))

#4. Calculating bias error
bias = np.mean(analyzed_data-obs)

# Printing results
print("Observations: ", obs)
print("Model_output: ", model_output)
print("Weight_obs: ", W_obs)
print("Weight_model: ", W_mod)
print("Analyzed_data: ", analyzed_data)
print("RMSE: ", rmse)
print("Bias error: ", bias)
```

## DISPLAY OF CALACULATED RESULTS

```python
print(pd.DataFrame(
    { "OBSERVATIONS(°C)": obs,
      "MODEL OUTPUT(°C)": model_output,
      "ANALYZED DATA(°C)": analyzed_data,
      "RMSE": rmse,
      "BIAS ERROR": bias,
    }
))
```

## PLOTTING COMPARISION DATASET

```python
#comparing analyzed data with Observation data
x = ([1, 2, 3, 4, 5])
obs = ([15.2, 16.1, 14.5, 15.8, 25.0])
model_output = ([14.8, 15.5, 14.0, 16.0, 25.3])
Analyzed_data =([15.0, 15.8, 14.3, 15.9, 25.1])

# plotting of figure
fig = plt.subplots(figsize = (10,5))
plt.plot(x,obs,label="observation", color= "red", marker = ".")
plt.plot(x,model_output, label = "model", color = "blue", marker ="^")
plt.plot(x,Analyzed_data, label = "Analysis", color = "green", marker = "o")
plt.legend()
plt.grid(True)
plt.title("COMPARISON ANALYSIS BETWEEN DATASETS", fontweight = "bold")
plt.xlabel("Number of counts", fontweight = "bold")
plt.ylabel("Dataset(°C)", fontweight = "bold")
plt.show()
plt.savefig("new3.png")
```
