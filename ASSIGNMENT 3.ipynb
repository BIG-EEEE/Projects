## QUESTION 1
## Use the euler's method with step size  h= 0.2  to find approximate values of the solution of the initial value problem y=2x^3 + x^2. Start at (x,y) = (0,1) , 0<=x<=2. Compare these approximate values with the values of the exact solution. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Differential equation dy/dx 
def f(x, y):
    return 6*x**2+2*x

# Exact solution
def exact_solution(x):
    return 2*x**3+x**2

# Euler's method
def euler_method(f, x0, y0, h, x_end):
    n = int((x_end - x0) / h)  # Number of steps
    x_values = np.linspace(x0, x_end, n + 1)
    y_values = np.zeros(n + 1)
    y_values[0] = y0
    
    for i in range(n):
        y_values[i+1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    return x_values, y_values

# Initial conditions
x0 = 0
y0 = 1
h = 0.4
x_end = 2

# Compute values using Euler's method
x_values, y_approx = euler_method(f, x0, y0, h, x_end)



# Compute exact values
y_exact = exact_solution(x_values)
y_exact[0] = 1

# Calculate absolute
Absolute = np.abs(y_exact - y_approx)
# Calculate relative error
rel_error = ((Absolute)/(y_exact))*100



# Display results
print("x values:", x_values)
print("Approximate y values:", y_approx)
print("Exact y values:", y_exact)
print("Relative error:", rel_error)
print("Absolute error:", Absolute)
```

```python
pd.DataFrame(
    {
        "X":x_values,
        "Exact Values":y_exact,
        "Approximated Values":y_approx,
        "Absolute Error":Absolute,
        "Relative Error":rel_error,
    }
)
```

```python

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_approx, 'bo-', label="Euler's Method Approximation")
plt.plot(x_values, y_exact, 'r-', label="Exact Solution")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Comparison of Euler's Method and Exact Solution")
plt.legend()
plt.grid(True)
plt.show()
```
